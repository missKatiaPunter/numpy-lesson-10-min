### Definition

**NumPy Array** is one of the key data structures for Data Scientists. It:

* is more efficient than a list (faster, more Pythonic, and has less memory footprint)
* homogenous
* allows operating on a set of values at one time (vectorized arithmetic operations)
* allows masking and filtering on a set of values without using loops or filter

#### Part 1. Creating an Numpy array and doing some basic vector math on it

```
example_list = [2,3,4,5] #creates a Python list
type(example_list)
```

```
import numpy as np
np_list = np.array(example_list) # turns a list into a numpy array
type(np_list)
```

```
np_list_squared = np_list ** 2 # squares each element of the vector fast
np_list_squared
```

#### Part 2. NumPy filter array

```
from numpy import array

np_array_to_be_filtered = array(range(0,20)) # creates a numpy array of numbers 0 to 20

keep_evens_filter = np_array_to_be_filtered % 2 == 0 # creates a filter array to filter out odds

np_array_to_be_filtered[keep_evens_filter] # applies filter array
```

```
np_array_to_be_filtered
```

```
keep_evens_filter # shows what a filter array looks like
```

#### Part 3. Evaluating np array's efficiency

```
import time
start_time = time.perf_counter()
evens = []
for x in range(0, len(array(range(0,2000000)))):
    if x % 2 == 0:
        evens.append(x)
evens
end_time = time.perf_counter()
print(end_time - start_time, "seconds")
```

```
import time
start_time = time.perf_counter()
np_array_efficient = array(range(0,2000000))
evens_filter = np_array_efficient % 2 == 0
evens2 = np_array_efficient[evens_filter]
end_time = time.perf_counter()
print(end_time - start_time, "seconds")
```
